```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N_SAMPLES = 500
```

```python
input_range = np.linspace(0,2*np.pi,N_SAMPLES, dtype=np.float32)
input_range
```

```python
signal = np.sin(input_range)
```

```python
noise = np.random.normal(0,1,N_SAMPLES)
```

```python
assert noise.shape == input_range.shape
```

```python
noisy_signal = signal + noise/5
```

```python
plt.plot(input_range, noisy_signal,'b.',alpha=0.5)
plt.plot(input_range, signal,'r-',linewidth=4)
plt.legend(['Sensor readings', 'Truth'])
plt.xlabel('Time')
plt.ylabel('Value of the thing we care about')
plt.show()
```

```python
del signal
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
filtered = lowess(noisy_signal,input_range,frac=0.1)
```

```python
plt.plot(input_range, noisy_signal,'b.',alpha=0.5)
plt.plot(filtered[:,0], filtered[:,1],'r-',linewidth=4)
plt.legend(['Sensor readings', 'Reconstructed signal'])
plt.show()
```

```python
def np_summary():
    data = np.load('monthdata.npz')
    totals = data["totals"]
    counts = data["counts"]
    sum1 = totals.sum(axis=1)
    lowest = sum1.argmin(axis=0)
    print("Row with lowest total precipitation: n",lowest)
    
    
    avgMonthRain = totals.sum(axis=0)/counts.sum(axis=0)
    print("Average precipitation in each month: ",avgMonthRain)
    
    avgCityRain = totals.sum(axis=1)/counts.sum(axis=1)
    print("Average precipitation in each city: ",avgCityRain)
    
    reshapedTotals = np.reshape(totals,(4 * totals.shape[0],3))
    totalRainQuarter = reshapedTotals.sum(axis=1)
    totalRainQuarter = np.reshape(totalRainQuarter,(totalRainQuarter.shape[0]//4,4))
    print("Quarterly precipitation totals: ", totalRainQuarter)
    #Row with lowest total precipitation:
# 8
# Average precipitation in each month:
# [27.77978339 30.42629482 29.41007194 17.96654275 21.34456929 20.69498069
#  24.97718631 19.85661765 24.06563707 44.68441065 34.61568627 32.36679537]
# Average precipitation in each city:
# [47.77859779 14.33333333 39.91232877 41.44505495 23.4099723  23.68144044
#  17.52197802 36.52222222  6.15426997]
# Quarterly precipitation totals:
# [[5450 1408 1466 4624]
#  [ 189 1339 3148  527]
#  [3120 3357 3386 4705]
#  [4416 3321 2024 5325]
#  [2024 1498 1721 3208]
#  [1786 1809 2557 2397]
#  [1583 1296 1729 1770]
#  [4602 1340 1250 5956]
#  [ 338  524  922  450]]

    
```
