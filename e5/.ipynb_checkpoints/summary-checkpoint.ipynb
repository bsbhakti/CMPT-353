```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


file_names = ['data-1.csv','data-2.csv','data-3.csv','data-4.csv','data-5.csv','data-6.csv']


for file_name in file_names:
    file = pd.read_csv(file_name)
    print(file_name)
    print("Mean ",file.mean())
    print("Corr ", file.corr())
    print("Max ",file.max())
    print("Min ",file.min())
#     plt.scatter(file["x"],file["y"])
#     plt.show()
    if(file_name == file_names[0]):
        print(f"Desc: The data has almost perfect positive linear relationship.")
    elif (file_name == file_names[1]):
        print(f"Desc: The data has almost perfect positive linear relationship with some outliers.")
    elif(file_name == file_names[2]):
        print(f"Desc: The data a bunch of clusters arranged with x and y being positively related..")
    elif(file_name == file_names[3]):
        print(f"Desc: The data forms the logo of SFU with no linear relationship")
    elif(file_name == file_names[5]):
        print(f"Desc: The data is a line with slope 1 with some outliers.")
    else:
        print(f"Desc: The data is some parallel lines in the range [-20,20]")
    print("\n")





   
```
