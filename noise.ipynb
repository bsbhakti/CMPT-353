```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Creation

```python
n_samples = 500
x = np.linspace(-10, 10, n_samples, dtype=np.float64)
data = pd.DataFrame()
data['x'] = x
data['y1'] = (x * x - 10
        + np.random.normal(0, 15, n_samples))
data['y2'] = (2 * x
        + np.random.normal(0, 10, n_samples))
```

```python
plt.plot(data['x'], data['y1'], 'b.')
plt.plot(data['x'], data['y2'], 'r.');
```

## LOESS Filtering

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
f1 = lowess(data['y1'], data['x'], frac=0.3)
f2 = lowess(data['y2'], data['x'], frac=0.2)
```

```python
plt.plot(data['x'], data['y1'], 'b.')
plt.plot(data['x'], f1[:, 1], 'b-');
```

```python
plt.plot(data['x'], data['y2'], 'r.')
plt.plot(data['x'], f2[:, 1], 'r-');
```

## Kalman Filtering

```python
from pykalman import KalmanFilter
```

```python
observations = data[['y1','y2']]
```

```python
initial_value_guess = observations.iloc[0]
```

```python
observation_covariance = np.diag([15, 10]) ** 2
```

```python
transition_covariance = np.diag([15, 10]) ** 2
```

```python
transition_matrix = [[1, 1], [0, 1]]
```

```python
kf = KalmanFilter(
    initial_state_mean=initial_value_guess,
    initial_state_covariance=observation_covariance,
    observation_covariance=observation_covariance,
    transition_covariance=transition_covariance,
    transition_matrices=transition_matrix
)
pred_state, state_cov = kf.smooth(observations)
```

```python
plt.plot(data['x'], data['y1'], 'b.')
plt.plot(data['x'], pred_state[:, 0], 'b-');
```

```python
plt.plot(data['x'], data['y2'], 'r.')
plt.plot(data['x'], pred_state[:, 1], 'r-');
```
